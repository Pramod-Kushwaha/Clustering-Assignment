# **Theoretical Questions**

1. What is unsupervised learning in the context of machine learning?
   - Unsupervised learning is a type of machine learning where the algorithm is provided with unlabeled data and must find patterns or relationships within it without having any explicit output labels. In clustering, a common unsupervised task, the goal is to group similar data points together based on their features so that points in the same group (cluster) are more alike than those in different groups.

2. How does the K-Means clustering algorithm work?
   - K-Means is an iterative clustering algorithm that partitions data into k distinct clusters. It begins by initializing k centroids randomly. Each data point is then assigned to the nearest centroid, forming clusters. Next, the centroids are recalculated as the mean of all points in the cluster. This process of assignment and updating centroids repeats until the centroids no longer change significantly or a maximum number of iterations is reached.

3. Explain the concept of a dendrogram in hierarchical clustering.
   - A dendrogram is a tree-like diagram that illustrates the arrangement of clusters produced by hierarchical clustering. It shows how individual data points are merged into clusters step by step. The vertical axis represents the distance or dissimilarity at which clusters are joined, and by cutting the dendrogram at a certain height, you can choose the desired number of clusters.

4. What is the main difference between K-Means and Hierarchical Clustering?
   - The main difference is in their clustering approach. K-Means is a partition-based method that requires the number of clusters k as input and optimizes the assignment to minimize within-cluster variance. Hierarchical clustering builds a hierarchy of clusters either by successively merging smaller clusters (agglomerative) or dividing larger clusters (divisive), and does not require specifying the number of clusters upfront.

5. What are the advantages of DBSCAN over K-Means?
   - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means:

    - It can identify clusters of arbitrary shape, not just spherical clusters.

    - It does not require specifying the number of clusters in advance.

    - It can detect and label noise and outlier points.

    - It handles clusters with varying densities better than K-Means.

6. When would you use Silhouette Score in clustering?
   - The Silhouette Score is used to evaluate the quality of clustering by measuring how similar a point is to its own cluster compared to other clusters. It is typically used to select the optimal number of clusters or compare different clustering solutions. A higher silhouette score indicates better-defined and well-separated clusters.

7. What are the limitations of Hierarchical Clustering?
   - Hierarchical Clustering can be computationally expensive for large datasets, as it has high time and space complexity. It is sensitive to noise and outliers. Once a merge or split is made, it cannot be undone, which may lead to suboptimal clustering. Also, it does not scale well to very large datasets.

8. Why is feature scaling important in clustering algorithms like K-Means?
   - Feature scaling ensures that all features contribute equally to distance calculations in clustering algorithms. In K-Means, which uses Euclidean distance, variables with larger ranges can dominate the calculation, leading to biased clustering. Scaling standardizes features so each has equal influence on cluster formation.

9. How does DBSCAN identify noise points?
   - DBSCAN considers a point as noise (or an outlier) if it does not have enough neighboring points within a specified radius (ε). Such points do not belong to any cluster, as they fail to satisfy the minimum number of neighbors (minPts) required to form a dense region.

10. Define inertia in the context of K-Means.
    - Inertia refers to the sum of squared distances between each data point and its assigned cluster centroid in K-Means. Lower inertia means that data points are closer to their centroids, indicating tighter and more cohesive clusters.

11. What is the elbow method in K-Means clustering?
    - The elbow method helps determine the optimal number of clusters in K-Means. It involves plotting inertia against various values of k. The point where the rate of decrease in inertia sharply slows (forming an “elbow” shape) suggests a suitable number of clusters.

12. Describe the concept of "density" in DBSCAN.
    - In DBSCAN, "density" refers to the number of data points within a specified radius (ε) around a point. A region is considered dense if it contains at least a minimum number of points (minPts), which can form a cluster.

13. Can hierarchical clustering be used on categorical data?
    - Yes, hierarchical clustering can be applied to categorical data, but it requires an appropriate distance metric (such as Hamming distance or Jaccard distance) that can handle categorical variables, since traditional methods like Euclidean distance are not suitable.

14. What does a negative Silhouette Score indicate?
    - A negative Silhouette Score means that a data point has been assigned to the wrong cluster, as it is closer on average to points in another cluster than to those in its own cluster. It indicates poor clustering quality.

15. Explain the term "linkage criteria" in hierarchical clustering.
    - Linkage criteria determine how the distance between clusters is computed during hierarchical clustering. Common linkage methods include single linkage (minimum distance), complete linkage (maximum distance), average linkage (mean distance), and ward linkage (minimizing variance).

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
    - K-Means assumes all clusters have similar size, shape, and density. When clusters vary in size or density, K-Means may split large or dense clusters or merge small or sparse clusters incorrectly, leading to poor clustering results.

17. What are the core parameters in DBSCAN, and how do they influence clustering?
    - DBSCAN has two main parameters:

      - ε (epsilon): The radius for neighborhood search.

      - minPts: The minimum number of points required to form a dense region.

     These parameters determine how clusters are created and which points are labeled as noise.

18. How does K-Means++ improve upon standard K-Means initialization?
    - K-Means++ improves K-Means by initializing cluster centroids in a smart way: it selects the first centroid randomly and then picks subsequent centroids with a probability proportional to their squared distance from the nearest existing centroid. This reduces the likelihood of poor clustering and speeds up convergence.

19. What is agglomerative clustering?
    - Agglomerative clustering is a type of hierarchical clustering that starts with each data point as its own cluster and then repeatedly merges the closest pairs of clusters until all points belong to a single cluster or a stopping criterion is met.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
    - While inertia only measures cluster cohesion (how close points are within a cluster), Silhouette Score also considers separation (how distinct clusters are from one another). This dual perspective provides a more comprehensive evaluation of clustering quality.

```python
#21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=400, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
y_pred = kmeans.fit_predict(X)

# Visualization
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.title("K-Means Clustering on Blobs (4 Centers)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
#23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=300, noise=0.09, random_state=0)
db = DBSCAN(eps=0.25, min_samples=5)
labels = db.fit_predict(X)
plt.scatter(X[labels != -1, 0], X[labels != -1, 1], c=labels[labels != -1], cmap='Set1', label='Clusters')
plt.scatter(X[labels == -1, 0], X[labels == -1, 1], c='black', marker='x', s=60, label='Outliers')
plt.title("DBSCAN on make_moons with Outliers Highlighted")
plt.legend()
plt.show()
```

```python
#24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

wine = load_wine()
X = wine.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)
unique, counts = np.unique(labels, return_counts=True)
for cluster, count in zip(unique, counts):
    print(f"Cluster {cluster}: {count} samples")
```

```python
#29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=400, centers=4, random_state=42)
silhouette_scores = []
ks = range(2, 6)
for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)
plt.bar(ks, silhouette_scores, color='skyblue')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for different k in K-Means')
plt.show()
```

```python
#31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Synthetic overlapping data
X, y = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
labels = kmeans.predict(X)

# Plot decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05), np.arange(y_min, y_max, 0.05))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.2)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=30, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=250, c='red', marker='X')
plt.title("K-Means with Decision Boundaries (Overlapping Data)")
plt.show()
```

```python
#34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = load_breast_cancer().data
X_scaled = StandardScaler().fit_transform(X)
ks = range(2, 7)
inertias = []
for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
plt.plot(ks, inertias, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('K-Means Inertia for Different k (Breast Cancer)')
plt.show()
```
